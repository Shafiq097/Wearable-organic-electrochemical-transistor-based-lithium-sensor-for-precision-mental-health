```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame, Panel
pd.set_option('display.max_rows',15) # this limit maximum numbers of rows
# import libs
import matplotlib
import os.path
import matplotlib.pyplot as plt
import numpy as np
from numpy import savetxt
import matplotlib.ticker as ticker
from matplotlib import rcParams
from matplotlib import rc
import matplotlib.gridspec as gridspec
matplotlib.font_manager._rebuild()
import matplotlib.ticker as ticker
from scipy.optimize import curve_fit
from scipy import mean
import scipy
import scipy.optimize
rc('font',**{'family':'sans-serif','sans-serif':['PT Sans'],'size': 55})
colors = ['#4E79A7', '#F28E2B', '#E15759', '#59A14E', '#EDC949', '#AF7AA1', '#FF9DA7', '#9C755F', '#BAB0AC', '#76B7B2']
```

```python
# open file
data = np.genfromtxt('Figure_4C.csv', delimiter=',', skip_footer=0, skip_header=2)

data1 = data[0:6,1]
data2 = data[0:6,3]
data3 = data[0:6,5]

Con_plot=data[:,6]
Id_plot_1 = data[:,7]

con=[10,5,2,1,0.5,0.1]
rI_avg=(data1+data2+data3)/3
```

```python
#standard deviation
rI_sd=np.zeros(len(rI_avg))

s=0
for i in range(len(rI_avg)):
    s=(data1[i]-rI_avg[i])**2+(data2[i]-rI_avg[i])**2+(data3[i]-rI_avg[i])**2
    
    rI_sd[i]=(s/(3))**0.5 
rI_sd2=rI_sd
rI_sd=rI_sd*.5
```

```python
#Figure 4C
fig1, ax = plt.subplots(1, figsize=(5,5))
plt.errorbar(con, np.abs(rI_avg), yerr=rI_sd, fmt='o',markersize=10,solid_capstyle='projecting', capsize=2,color=colors[9])
plt.errorbar(con, np.abs(rI_avg)+0.0005, fmt='o',solid_capstyle='projecting', capsize=2,ms=1.7,color='white',alpha=0.7)
plt.plot(Con_plot,Id_plot_1,linewidth=2,linestyle='dashed', color='gray', alpha = 1,label='n1')

font_style = {
    'family': 'PT sans',
    'color': 'black',
#    'weight': 'bold',
    'size': 24,
}


plt.xscale('log')
plt.xlabel('Concentration (mM) ' ,fontdict=font_style)
plt.ylabel(r'$|\Delta$ $I_{ds}/I_0|$', fontdict=font_style)


plt.tick_params(direction='in', right=False, top=False)
plt.tick_params(labelsize=22)
plt.tick_params(direction='in', which='major', length=5, bottom=True, left=True)
plt.tick_params(direction='in', which='minor', length=2, bottom=False, left=False)
ax.tick_params(axis='x', pad=9)
ax.tick_params(axis='y', pad=9)

plt.ylim([0,0.06])
#plt.xlim([1,160])
ax.set_yticks([0.0,0.02,0.04,0.06])
```
