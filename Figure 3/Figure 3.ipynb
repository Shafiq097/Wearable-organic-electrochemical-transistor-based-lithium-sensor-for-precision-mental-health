```python
# open file
data_3b = np.genfromtxt('Figure_3B.csv', delimiter=',', skip_footer=0, skip_header=2)
```

```python
Con=data_3b[:,0]
data1=data_3b[:,1]
data2=data_3b[:,3]
data3=data_3b[:,5]
Con_fit=data_3b[:,6]
data_fit=data_3b[:,7]

rI_avg=(data1+data2+data3)/3

#standard deviation
rI_sd=np.zeros(len(rI_avg))

s=0
for i in range(len(rI_avg)):
    s=(data1[i]-rI_avg[i])**2+(data2[i]-rI_avg[i])**2+(data3[i]-rI_avg[i])**2
    
    rI_sd[i]=(s/(3))**0.5 
rI_sd2=rI_sd
rI_sd=rI_sd*.5
```

```python
#Figure 3B
fig1, ax = plt.subplots(1, figsize=(5,5))
plt.errorbar(Con, np.abs(rI_avg), yerr=rI_sd, fmt='o',markersize=10,solid_capstyle='projecting', capsize=2,color='midnightblue')
plt.errorbar(Con, np.abs(rI_avg)+0.001, fmt='o',solid_capstyle='projecting', capsize=2,ms=1.7,color='white',alpha=0.7)
plt.plot(Con_fit,data_fit,linewidth=2,linestyle='dashed', color='red', alpha = 1,label='n1')

font_style = {
    'family': 'PT sans',
    'color': 'black',
#    'weight': 'bold',
    'size': 24,
}

plt.xscale('log')
plt.xlabel('Concentration (mM) ' ,fontdict=font_style)
plt.ylabel(r'$|\Delta$ $I_{ds}/I_0|$', fontdict=font_style)

plt.tick_params(direction='in', right=False, top=False)
plt.tick_params(labelsize=22)
plt.tick_params(direction='in', which='major', length=5, bottom=True, left=True)
plt.tick_params(direction='in', which='minor', length=2, bottom=False, left=False)
ax.tick_params(axis='x', pad=9)
ax.tick_params(axis='y', pad=9)
plt.minorticks_on()


ax.set_yticks([0.05,.15,.25,.35])
ax.set_xticks([0.001, 0.01,.1,1, 10])
```
