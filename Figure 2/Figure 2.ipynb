```python
import pandas as pd
import numpy as np
#from pandas import Series, DataFrame, Panel
#pd.set_option('display.max_rows',15) # this limit maximum numbers of rows
# import libs
import matplotlib
import os.path
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker
from matplotlib import rcParams
from matplotlib import rc
import matplotlib.gridspec as gridspec
#matplotlib.font_manager._rebuild()
import matplotlib.ticker as ticker
from scipy.optimize import curve_fit
import scipy
import scipy.optimize
from matplotlib.legend_handler import HandlerTuple
from scipy.optimize import curve_fit
rc('font',**{'family':'sans-serif','sans-serif':['PT Sans'],'size': 55})
colors = ['#4E79A7', '#F28E2B', '#E15759', '#59A14E', '#EDC949', '#AF7AA1', '#FF9DA7', '#9C755F', '#BAB0AC', '#76B7B2']
```

```python
# open file
fig_2b = np.genfromtxt('Figure_2B.csv', delimiter=',', skip_footer=0, skip_header=1)
```

```python
# Figure 2B
from scipy import mean
Vgs1 = fig_2b[:,0]
Vds1 = fig_2b[:,1]
Ids1 = fig_2b[:,2]/1e-3 #Id in mA

fig1, ax = plt.subplots(1, figsize=(5,5))


n_vg=12; #number of Vgs point
n_vd=80; #number of Vds point
Vds=np.zeros(80)
Ids=np.zeros(80)
n=n_vg*n_vd
k=0
for j in range(n_vg):
    
    Vds=np.zeros(n_vd)
    Ids=np.zeros(n_vd)
    for i in range(n_vd):
        Vds[i]=Vds1[k]
        Ids[i]=Ids1[k]
        k=k+1
        
    plt.plot(Vds[0:78],Ids[0:78],linewidth=2, color=colors[0],alpha =1,label='n1')
 
font_style = {
    'family': 'PT sans',
    'color': 'black',
#    'weight': 'bold',
    'size': 24,
}


plt.tick_params(direction='in', right=False, top=False)
plt.tick_params(labelsize=22)
plt.tick_params(direction='in', which='major', length=5, bottom=True, left=True)
plt.tick_params(direction='in', which='minor', length=2, bottom=True, left=True)
ax.tick_params(axis='x', pad=7)
ax.tick_params(axis='y', pad=7)
plt.ylim([-8,0])
plt.xlim([-.8,0])
ax.set_yticks([0,-2,-4,-6,-8])
ax.set_xticks([-.8, -.6, -.4, -.2, 0])

plt.xlabel(r'V$_{ds}$  (V)', fontdict=font_style)
plt.ylabel(r'I$_{ds}$  (mA)', fontdict=font_style)

```
